## Web Scraping Code

```python
import asyncio
from typing import List
import httpx
# from walkscore_api import WalkScore
from bs4 import BeautifulSoup

import json
from parsel import Selector
import pandas as pd
```

```python
async def scrape_properties_links(urls: List[str], client,base=None):

    to_scrape = [client.get(url) for url in urls]
    
    results = []
    i=0
    for response in asyncio.as_completed(to_scrape):
        response = await response
        assert response.status_code == 200, "request has been blocked"
        soup = BeautifulSoup(response.text, 'html.parser')
        # Find the section with class "bh-content-component"
        section = soup.find('section', class_='bh-content-component')
        if section:
            # Find the <ul> element within the section
            ul_tag = section.find('ul')
            
            if ul_tag:

                # Find all <a> tags within the <ul> element
                for a_tag in ul_tag.find_all('a'):
                    url = a_tag.get('href')
                    if(base==None):
                        results.append(urls[i]+url)
                    else :
                        results.append(base+url)
                i+=1
    
    return results
```

```python
client = httpx.AsyncClient(
    # enable http2
    http2=True,
    # add basic browser like headers to prevent being blocked
    headers={
        "accept-language": "en-US,en;q=0.9",
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36",
        "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
        "accept-language": "en-US;en;q=0.9",
        "accept-encoding": "gzip, deflate, br",
    },
)

urls = ['https://www.zillow.com/browse/homes/']
```

```python
base="https://www.zillow.com"
results = await scrape_properties_links(urls, client, base)
df=pd.DataFrame(results)
urls = []

# Iterate through the DataFrame rows
for index, row in df.iterrows():
    urls.append(row[0])

# Print or use the list of URLs as needed
print(urls)
```

```python
async def scrape_properties_batched(urls: List[str], client,init_val=0, batch_size=100, timeout=30):
    # results = []
    for i in range(0, len(urls), batch_size):
        batch = urls[i:min(i+batch_size, len(urls))]  # Ensure no out of bounds index
        batch_results = await scrape_properties(batch, client)
        # results.extend(batch_results)
        df=pd.DataFrame(batch_results)
        df.to_json(f"data/ListingInfoFinal/{init_val+i}.json")
        await asyncio.sleep(timeout)
    # return results

# Usage
results = await scrape_properties_batched(urlsnew1[2000:], client,init_val=2000)
```
