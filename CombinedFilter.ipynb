```python
import pandas as pd
import json

# Read JSON file
with open('combined_filtered_data.json') as f:
    data = json.load(f)

# Convert JSON data to DataFrame
df = pd.json_normalize(data)

# Save DataFrame to CSV
df.to_csv('data.csv', index=False)

print("CSV file created successfully.")
```
