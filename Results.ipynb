```python
import pandas as pd
import matplotlib.pyplot as plt
import math
df = pd.read_csv("./Results.csv")
df
```

```python
print(df['Mean ROUGE Score IC'].mean())
print(df['Mean ROUGE Score WOIC'].mean())
print(df['Mean ROUGE Score CC'].mean())
```

```python
print(math.sqrt(df['Variance in ROUGE Score IC'].mean()))
print(math.sqrt(df['Variance in ROUGE Score WOIC'].mean()))
print(math.sqrt(df['Variance in ROUGE Score CC'].mean()))
```

```python
plt.figure(figsize=(12,7))
plt.plot(df.index, df['Mean ROUGE Score IC'], label='Mean ROUGE Score with Image Captions', color='b')  # Blue color for Category1
plt.plot(df.index, df['Mean ROUGE Score WOIC'], label='Mean ROUGE score without Image Captions', color='g')  # Green color for Category2
plt.plot(df.index, df['Mean ROUGE Score CC'], label='Mean ROUGE Score with Only Clustering', color='r')  # Red color for Category3

plt.xlabel('Index')
plt.ylabel('ROUGE Score')
plt.ylim(0,0.06)
plt.title('Line Chart for the ROUGE Score for each algorithm')
plt.legend()
plt.grid(True)
plt.savefig("output.png")
plt.show()
```
