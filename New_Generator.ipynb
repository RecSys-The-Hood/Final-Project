```python
import pandas as pd
import json
```

```python
df = pd.read_csv("Final_Combined_Dataset.csv")
df.columns
```

```python
states = ["GA", "MA", "IL", "TX", "MI", "CA", "FL", "NY", "PA", "AZ", "CA", "DC"]
df = df[df['address.state'].isin(states)]
df
```

```python
groups = df.groupby(by=['address.state'])
groups.count()['zpid']
```

```python
list_of_dfs = [group for _, group in groups]
states = [state for state, _ in groups]
list_of_dfs[0].head(5)
```

```python
rates = {}
schools = {}
leisure = {}
shops = {}
transit = {}

schools_std = {}
leisure_std = {}
shops_std = {}
transit_std = {}

for i in range(0, len(list_of_dfs)):
# for df in list_of_dfs:
    df1 = list_of_dfs[i]
    rates[states[i]] = df1['price'].sum() / df1['livingArea'].sum()
    leisure[states[i]] = df1['leisure_within_5km'].mean()
    schools[states[i]] = df1['schools_within_5km'].mean()
    shops[states[i]] = df1['shops_within_5km'].mean()
    transit[states[i]] = df1['transit_within_2km'].mean()
    leisure_std[states[i]] = df1['leisure_within_5km'].std()
    schools_std[states[i]] = df1['schools_within_5km'].std()
    shops_std[states[i]] = df1['shops_within_5km'].std()
    transit_std[states[i]] = df1['transit_within_2km'].std()

print(rates)
print(schools)
print(leisure)
print(shops)
print(transit)
```

```python
import json

with open('rates.json', 'w') as json_file:
    json.dump(rates, json_file, indent=4)

with open('schools.json', 'w') as json_file:
    json.dump(schools, json_file, indent=4)

with open('transit.json', 'w') as json_file:
    json.dump(transit, json_file, indent=4)

with open('shops.json', 'w') as json_file:
    json.dump(shops, json_file, indent=4)

with open('leisure.json', 'w') as json_file:
    json.dump(leisure, json_file, indent=4)

with open('schools_std.json', 'w') as json_file:
    json.dump(schools_std, json_file, indent=4)

with open('transit_std.json', 'w') as json_file:
    json.dump(transit_std, json_file, indent=4)

with open('shops_std.json', 'w') as json_file:
    json.dump(shops_std, json_file, indent=4)

with open('leisure_std.json', 'w') as json_file:
    json.dump(leisure_std, json_file, indent=4)
```
