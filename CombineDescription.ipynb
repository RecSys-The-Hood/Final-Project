```python
import pandas as pd
import numpy as np
import torch

from transformers import pipeline
```

```python
torch.cuda.empty_cache()
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
df = pd.read_csv('Final_dataset.csv')
df.columns
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
def summarize_text(text):
    summary = summarizer(text, max_length=80, min_length=50, do_sample=False)  # Adjust min/max lengths
    return summary[0]['summary_text']
```

```python
df['final_description'] = df['description']+df['concatenated_desc']

df['summarized_description'] = df['final_description'].apply(summarize_text)
```
