```python
import pandas as pd
import ast
```

```python
df = pd.read_csv("./ListingsAcrossStates.csv")
print(df.columns)
```

```python
cities = ['Atlanta', 'Boston', 'Chicago', 'Dallas', 'Detroit', 'Houston', 'Las Vegas', 'Los Angeles', 'Miami', 'New York', 'Philadelphia', 'Phoenix', 'San Francisco', 'San Jose', 'Washington']
df = df[df['addressCity'].isin(cities)]
df.groupby(by=['addressCity']).count()['zpid']
```

```python
df
```

```python
for col in df.columns:
    print(col+" "+str(df[col].count()))
```

```python
df.drop(columns=['isPropertyResultCDP', 'providerListingId', 'builderName', 'availabilityDate', 'streetViewMetadataURL', 'streetViewURL', 'openHouseStartDate', 'openHouseEndDate'], inplace=True)
df.columns
```

```python
are_equal = (df['zpid'] == df['id']).all()
df.drop(columns=['id'], inplace=True)
```

```python
df.dropna(subset=['baths', 'area', 'carouselPhotos'], inplace=True)
df
```

```python
def extract_value(string, key):
    dictionary = ast.literal_eval(string)  # Convert string to dictionary
    return dictionary.get(key) 
```

```python
def parse_list(string):
    return [x['url'] for x in ast.literal_eval(string)]
```

```python
extracted_lats = df['latLong'].apply(lambda x: extract_value(x, 'latitude'))
extracted_longs = df['latLong'].apply(lambda x: extract_value(x, 'longitude'))
# Convert the list of values to a new series
df['latitude'] = pd.Series(extracted_lats)
df['longitude'] = pd.Series(extracted_longs)
df.drop(columns=['latLong'], inplace=True)
```

```python
lists_column = df['carouselPhotos'].apply(parse_list)
sizes_column = lists_column.apply(len)

df['sizes'] = sizes_column
df['urls_list'] = lists_column

df = df[df['urls_list'].apply(len) > 5]
df.drop(columns=['carouselPhotos'], inplace=True)
```

```python
n = df.shape[0]
df_new = pd.DataFrame(index=range(n), data=df.values, columns=df.columns)
df = df_new
df
```

```python
df.to_csv("ListingInCities.csv", index=False)
```
