## Sentence Embedding Generator

```python
from sentence_transformers import SentenceTransformer
# from sentence_transformers.util import cos_sim
# from sklearn.decomposition import PCA
# from sklearn.decomposition import TruncatedSVD
import pandas as pd
import json
```

```python
model = SentenceTransformer("mixedbread-ai/mxbai-embed-large-v1")
```

```python
df = pd.read_csv("./combined_summary_data(in).csv")
df = df.head()
lst = list(df['summary'])
```

```python
embeddings = model.encode(lst)
# print(embeddings.shape)
# print(embeddings.dtype)
```

```python
embeddings = embeddings.astype(float)
data = {int(df.iloc[i]['zpid']): list(embeddings[i]) for i in range(0, df.shape[0])}
print(data)

with open('data.json', 'w') as file:
    json.dump(data, file)
```
